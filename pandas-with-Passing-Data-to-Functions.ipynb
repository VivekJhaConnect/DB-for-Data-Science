```python
import pandas as pd
import numpy as np

import sys
```

```python
def get_sales_total(df):
    sales_ser = df['SALES']
    print(f"Total Profit : {sales_ser.sum()}")
```

```python
sales_df =  pd.read_csv('..\datasets\sales_data_sample.csv', sep=",", encoding='latin1')
sales_df.head()
```

```python
get_sales_total(sales_df)
```

```python
def split_orderdate(df):
    def get_date(date):
        day, month, year = date.split('/')
        return pd.Series((day, month, year), index=['Day', 'Month', 'Year'])
    dates = df['ORDERDATE'].apply(get_date)
    df[dates.columns] = dates
    return df
```

```python
split_orderdate(sales_df).head()
```

```python
def create_QUANTITYORDERED_group(df):
    bins = [0, 30, 50, sys.maxsize]
    labels = ['<30', '30-50','>50']
    quant_group = pd.cut(df['QUANTITYORDERED'], bins=bins, labels=labels)
    df['Quant Order Group'] = quant_group
    return df
```

```python
create_QUANTITYORDERED_group(sales_df).head()
```

```python
sales_df.pipe(split_orderdate).pipe(create_QUANTITYORDERED_group).head()
```
